```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('data//14-Aug-2021-1.csv')
data2 = pd.read_csv('data//14-Aug-2021-2.csv')
data3 = pd.read_csv('data//14-Aug-2021-3.csv')
data4 = pd.read_csv('data//14-Aug-2021-4.csv')
```

```python
print(data1.shape)
print(data2.shape)
print(data3.shape)
print(data4.shape)
```

```python
R1 = pd.DataFrame()
R2 = pd.DataFrame()
R3 = pd.DataFrame()
R4 = pd.DataFrame()
```

```python
R1['Latitude']  = data1[' Latitude']
R1['Longitude'] = data1[' Longitude']
R1['KPL']       = data1['Trip average KPL(kpl)']
R1['EL']        = data1['Engine Load(%)']
R1['Speed']     = data1['Speed (OBD)(km/h)']
R1['Fuel_mass'] = (770 / 360) * data1['Fuel flow rate/hour(l/hr)']
R1['Trip Dist'] = data1['Trip Distance(km)']

R1['KPL'] = R1['KPL'].replace('-', 0)
R1['KPL'] = R1['KPL'].astype(float)
R1 = R1[R1['KPL'] > 0]

print(R1.head())
# print(R1.shape)
# print(R1.info())
```

```python
R2['Latitude']  = data2[' Latitude']
R2['Longitude'] = data2[' Longitude']
R2['KPL']       = data2['Trip average KPL(kpl)']
R2['EL']        = data2['Engine Load(%)']
R2['Speed']     = data2['Speed (OBD)(km/h)']
R2['Fuel_mass'] = (770 / 360) * data2['Fuel flow rate/hour(l/hr)']
R2['Trip Dist'] = data2['Trip Distance(km)']

R2['KPL'] = R2['KPL'].replace('-', 0)
R2['KPL'] = R2['KPL'].astype(float)
R2 = R2[R2['KPL'] > 0]

print(R2.head())
print(R2.shape)
print(R2.info())
```

```python
R3['Latitude']  = data3[' Latitude']
R3['Longitude'] = data3[' Longitude']
R3['KPL']       = data3['Trip average KPL(kpl)']
R3['EL']        = data3['Engine Load(%)']
R3['Speed']     = data3['Speed (OBD)(km/h)']
R3['Fuel_mass'] = (770 / 360) * data3['Fuel flow rate/hour(l/hr)']
R3['Trip Dist'] = data3['Trip Distance(km)']

R3['KPL'] = R3['KPL'].replace('-', 0)
R3['KPL'] = R3['KPL'].astype(float)
R3 = R3[R3['KPL'] > 0]

print(R3.head())
print(R3.shape)
print(R3.info())
```

```python
R4['Latitude'] = data4[' Latitude']
R4['Longitude'] = data4[' Longitude']
R4['KPL'] = data4['Trip average KPL(kpl)']
R4['EL'] = data4['Engine Load(%)']
R4['Speed'] = data4['Speed (OBD)(km/h)']
R4['Fuel_mass'] = (770 / 360) * data4['Fuel flow rate/hour(l/hr)']
R4['Trip Dist'] = data4['Trip Distance(km)']

R4['KPL'] = R4['KPL'].replace('-', 0)
R4['KPL'] = R4['KPL'].astype(float)
R4 = R4[R4['KPL'] > 0]

print(R4.head())
print(R4.shape)
print(R4.info())
```

```python
print(R1.shape)
print(R2.shape)
print(R3.shape)
print(R4.shape)
```

```python
x1 = np.arange(R1.shape[0])
x2 = np.arange(R2.shape[0])
x3 = np.arange(R3.shape[0])
x4 = np.arange(R4.shape[0])
```

```python
X1 = x1
X2 = x2 * ((len(x1)-1) / (len(x2)-1))
X3 = x3 * ((len(x1)-1) / (len(x3)-1))
X4 = x4 * ((len(x1)-1) / (len(x4)-1))
```

```python
Avg1 = round(np.mean(R1['KPL']),3)
Avg2 = round(np.mean(R2['KPL']),3)
Avg3 = round(np.mean(R3['KPL']),3)
Avg4 = round(np.mean(R4['KPL']),3)
```

### DataFrame for daily data

```python
data = {
    'Date': [data1['GPS Time'][1], data2['GPS Time'][1], data3['GPS Time'][1], data4['GPS Time'][1]],
    'Avg_Fuel':[Avg1, Avg2, Avg3, Avg4],
}
Daily_data_monitoring = pd.DataFrame.from_dict(data)
Daily_data_monitoring
```

```python
Daily_data_monitoring.to_csv('Daily_monitoring.csv')
```
